## NUMPY

NumPy es un paquete de Python. Significa 'Python numérico'. Es una biblioteca que consta de
objetos de matriz multidimensionales y una colección de rutinas para el procesamiento de matriz.

Con NumPy, un desarrollador puede realizar las siguientes operaciones:
- Operaciones matemáticas y lógicas en matrices.
- Transformaciones de Fourier y rutinas para la manipulación de formas.
- Operaciones relacionadas con el álgebra lineal. NumPy tiene funciones integradas para álgebra lineal
y generación de números aleatorios.

## OBJETO NDARRAY

El objeto más importante definido en NumPy es un tipo de matriz N-dimensional llamado ndarray.
Describe la colección de artículos del mismo tipo. Se puede acceder a los elementos de la colección
utilizando un índice de base cero.

Todos los elementos de un ndarray tienen el mismo tamaño de bloque en la memoria. Cada elemento en ndarray
es un objeto de tipo de datos (llamado dtype).

Cualquier elemento extraído del objeto ndarray (por corte) está representado por un objeto Python de uno
de tipos de matriz escalar. El siguiente diagrama muestra una relación entre ndarray, tipo de datos
objeto (dtype) y tipo escalar de matriz:

ndarray.PNG

Se puede construir una instancia de la clase ndarray mediante diferentes rutinas de creación de matrices descritas más adelante en el tutorial. El ndarray básico se crea usando una función de matriz en NumPy de la siguiente manera:

**numpy.array**

Crea un ndarray a partir de cualquier objeto que exponga la interfaz de matriz, o de cualquier método que
devuelve una matriz.

**numpy.array(object, dtype=None, copy=True, order=None, subok=False, ndmin=0)**

El constructor anterior toma los siguientes parámetros:

parámetros-ndarray.PNG


```python
# Creando un vector en Numpy

# Se importa la librería
import numpy as np

# Se crea un vector llamado 'a'
# Se utiliza el método 'array', el cual recibe una lista
a = np.array([1, 2, 3])

# Se imprime el vector
print (a)
```

```python
# Se crea una matriz de dos dimensiones
import numpy as np

# El método array recibe dos listas. La primera, es la primera fila
# La segunda lista, es la segunda fila de la matriz
a = np.array([[1, 2], [3, 4]])

# Se imprime la matriz
print (a)
```

```python
# Creación de un vector simple
import numpy as np
a = np.array([1, 2, 3, 4, 5])
print (a)

# NOTA: Como puede apreciarse, numpy ha creado un vector de una sola dimension 
```

```python
# Modificación dimensional
import numpy as np
a = np.array([1, 2, 3, 4, 5], ndmin = 2)  # Se pide a numpy que el objeto sea una matriz de 2 dimensiones
print(a)

# NOTA: Se crea una matriz de dos dimensiones, con una sola fila
#       Obsérvese los corchetes: [ [... primera fila ...] ]
#
# NOTA: No hay más filas, porque solo se entregó una lista
```

```python
# Cambiando el tipo de dato
# En el siguiente ejemplo, los datos NO son enteros, serán números complejos

# Parámetro dtype (tipo de dato)
import numpy as np
a = np.array([1, 2, 3], dtype = complex)  # El tipo de dato es un complejo (real + imaginario)
print(a)
```

## Atributos de Array

En este apartado, analizaremos los diversos atributos de matriz de NumPy.

## ndarray.shape

Este atributo de matriz devuelve una tupla que consta de la numeración de las dimensiones de la matriz. También se puede utilizar para cambiar el tamaño de la matriz.

```python
# En este ejemplo se calculan las dimensiones del array

import numpy as np
a = np.array([[1, 2, 3], [4, 5, 6]])
print (a)
print ()
print (a.shape)

# a.shape devuelve (2, 3). Esto significa: 2 filas, 3 columnas
```

```python
# En este ejemplo se cambia el tamaño del ndarray
import numpy as np
a = np.array([[1, 2, 3], [4, 5, 6]])

print ("Array original. Dos filas x Tres columnas")
print () 
print (a) 
print ()

a.shape = (3, 2)
print ("Array con su estructura modificada. Tres filas x Dos columnas")
print ()
print (a)
```

```python
# Función reshape para modificar la estructura de una matriz

import numpy as np

a = np.array([[1, 2, 3], [4, 5, 6]])
b = a.reshape(3, 2)
print (b)
```

## ndarray.ndim

Este atributo de matriz devuelve el número de dimensiones de la matriz.

```python
# una matriz de números espaciados uniformemente
import numpy as np

a = np.arange(24)   # El método 'arange' crea una lista de manera automática
                    # En este caso, los números del 0 al 23
print (a)
```

```python
# Esta es una matriz unidimensional
import numpy as np

a = np.arange(24)
print ("Valores para a: ")
print (a)
print ()
print ("Número de dimensiones de a: ")
print (a.ndim)    # ndim significa número de dimensiones
print ()

# Ahora reconfigurado
# Dos filas, cada una de ellas a su vez con 4 filas x 3 columnas
b = a.reshape(2,4,3)

print ("Valores para b: ")
print (b)
print ()
print ("Número de dimensiones de b: ")
print (b.ndim)
print ()
```

## numpy.itemsize

Este atributo de matriz devuelve la longitud de cada elemento de la matriz en bytes.

```python
# dtype del array es int8 (1 byte)
import numpy as np

x = np.array([1, 2, 3, 4, 5], dtype = np.int8)  # Tamaño de cada ítem (UN entero de 8 bits)
print (x.itemsize)
```

```python
# dtype del array es ahora un float32 (4 bytes)
import numpy as np

x = np.array([1, 2, 3, 4, 5], dtype = np.float32)
print (x.itemsize)
```

## Procedimientos para la creacion de arrays

Se puede construir un nuevo objeto ndarray mediante cualquiera de las siguientes rutinas de creación de matrices o utilizando un constructor ndarray de bajo nivel.

## numpy.empty

Crea una matriz no inicializada de forma y dtype especificados. Utiliza el siguiente constructor:

**numpy.empty(shape, dtype=float, order='C')**

El constructor toma los siguientes parámetros:

parámetros constructor.PNG

```python
# El siguiente código muestra un array vacío

import numpy as np
x = np.empty([3, 2], dtype=int)
print (x)

# NOTA: Aparecen elementos aleatorios que no han sido inicializados
```

## numpy.zeros

Devuelve una nueva matriz de tamaño especificado, llena de ceros.

**numpy.zeros(shape, dtype=float, order='C')**

El constructor toma los siguientes parámetros:

parametros-1.PNG

```python
# Array con cinco ceros. El tipo por defecto es float.
import numpy as np

x = np.zeros(5)
print (x)
```

```python
# Array con cinco ceros de tipo entero

import numpy as np
x = np.zeros((5,), dtype = np.int)  # La notación (5,) indica: vector de 5 posiciones
print (x)
```

```python
# custom type
import numpy as np

# La notación (2, 3) significa: matriz de 2 filas con 3 columnas
x = np.zeros((2, 3), dtype=[('x', 'i4'), ('y', 'i4'), ('z', 'i4')])

print (x)
```

## numpy.ones

Devuelve una nueva matriz de tamaño y tipo especificados, llena de unos.

**numpy.ones(shape, dtype=None, order='C')**

El constructor toma los siguientes parámetros:

parametros-3.PNG

```python
# Array de cinco unos. El tipo por defecto es float
import numpy as np
x = np.ones(5)
print (x)
```

```python
# Array de cinco unos de tipo entero
import numpy as np

# Dos filas, dos columnas
x = np.ones([2, 2], dtype = int)
print (x)
```

## numpy.asarray

Esta función es similar a numpy.array excepto por el hecho de que tiene menos parámetros. Este procedimiento es útil para convertir una secuencia de Python en ndarray.

## numpy.fromiter

Esta función crea un objeto ndarray a partir de cualquier objeto iterable. Una nueva matriz unidimensional
es devuelta por esta función.

**numpy.fromiter(iterable, dtype, count=-1)**

El constructor toma los siguientes parámetros:

parametros-1-a.PNG
parametros-1-b.PNG

Los siguientes ejemplos muestran cómo usar la función incorporada range () para devolver un objeto de lista.
Se utiliza un iterador de esta lista para formar un objeto ndarray.

```python
# Crea un objeto lista utilizando la función range

import numpy as np

lista = range(5)
print (lista)
```

```python
# Obtiene el objeto iterador a partir de la lista
import numpy as np

lista = range(5)
it = iter(lista)  # it es una variable que itera sobre la lista

# Se utiliza el iterador para crear un ndarray
x = np.fromiter(it, dtype = float)
print (x)
```

## Creando arrays a partir de rangos numéricos

En este apartado, veremos cómo crear una matriz a partir de rangos numéricos.

## numpy.arange

Esta función devuelve un objeto ndarray que contiene valores espaciados uniformemente dentro de un rango determinado. El formato de la función es el siguiente:

**numpy.arange(start, stop, step, dtype)**

El constructor toma los siguientes parámetros:

parametros-7.PNG

```python
# Crea un ndarray en el rango 0 hasta 4

import numpy as np

x = np.arange(5)
print (x)
```

```python
# Igual que el problema anterior, pero los datos son de tipo float

import numpy as np
# dtype set
x = np.arange(5, dtype = float)
print (x)
```

```python
# Crea un ndarray entre 10 y hasta antes de 20, de 2 en 2

import numpy as np

x = np.arange(10,20,2)
print (x)
```

## numpy.linspace

Esta función es similar a la función arange (). En esta función, en lugar del tamaño del paso, el número
de valores espaciados uniformemente entre el intervalo. El uso de esta función es como sigue:

**numpy.linspace(start, stop, num, endpoint, retstep, dtype)**

El constructor toma los siguientes parámetros:

1101.PNG
1111.PNG

```python
# EJEMPLOS:

# Se crea un ndarray entre 2.0 y 3.0. El número de puntos es de 5

import numpy as np

print (np.linspace(2.0, 3.0, num = 5))
```

```python
# No se imprime el último números del ndarray

import numpy as np

print (np.linspace(2.0, 3.0, num = 5, endpoint = False))
```

```python
# Se muestra el valor del incremento

print (np.linspace(2.0, 3.0, num = 5, retstep = True))
```
